```python
import matplotlib.pyplot as plt 
import numpy as np 
from mpl_toolkits.axes_grid1.inset_locator import mark_inset
from mpl_toolkits.axes_grid1.inset_locator import inset_axes
  
# create data 
x = np.arange(2) 

width = 0.12

ipmgnn_time = [0.0722, 0.0723, 0.0721, 0.0726]
scipy_time = [14.936164379119873, 15.417085886001587, 14.075480699539185, 18.322509288787842]

label = ['$\sf{WA}^{test}_{2000}$', '$\sf{WA}^{test}_{3000}$', 
         '$\sf{WA}^{test}_{4000}$', '$\sf{WA}^{test}_{5000}$']

color = ['#990011', '#DD2222', '#FF5555', '#FFAAAA']

labels = ['TELGEN', 'SciPy']

time = []
time.append(ipmgnn_time)
time.append(scipy_time)
time = np.array(time)

# Plot the main bar plot
fig, ax = plt.subplots(figsize=(6, 4))
width = 0.12

for i in range(len(ipmgnn_time)):
    ax.bar(x + 0.15 * (i - 0.35 * len(ipmgnn_time)), time[:, i], width, color=color[i], label=label[i])

ax.set_xticks(x)
ax.set_xticklabels(['TELGEN', 'SciPy'], fontsize=22, fontweight='bold') 
ax.set_yticks([0, 6, 12, 18])
ax.set_yticklabels(['0', '6', '12', '18'], fontsize=22)
ax.set_ylabel('Time (second)', fontsize=22, fontweight='bold')

# Create the inset for zooming in on the IPM-GNN bars
# ax_inset = inset_axes(ax, width="40%", height="40%", loc='upper left', borderpad=2)
ax_inset = ax.inset_axes((0.18, 0.1, 0.3, 0.4))
for i in range(len(ipmgnn_time)):
    ax_inset.bar(x + 0.15 * (i - 0.35 * len(ipmgnn_time)), time[:, i], width, color=color[i])

ax_inset.set_xlim(-0.3, 0.35)
ax_inset.set_ylim(0, 0.1)  # Adjust this range to focus on the small values
ax_inset.set_xticks([0])
# ax_inset.set_xticklabels(labels, fontsize=10, fontweight='bold')
ax_inset.set_yticks([0, 0.05, 0.1])
ax_inset.set_yticklabels(['0', '0.05', '0.1'], fontsize=22)

# Optionally, connect the inset to the main plot
mark_inset(ax, ax_inset, loc1=3, loc2=4, fc="none", ec="0.5")

# Add legend
ax.legend(loc=(0.001, 0.68), fontsize=22, handlelength=0.8, ncol=2, borderpad=0.1,
           handletextpad=0.2, labelspacing=0.1, columnspacing=0.1)

plt.show()
```
